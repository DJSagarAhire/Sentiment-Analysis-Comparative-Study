###Data Transformation

The following code transforms the documents into feature vectors depending on whether a particular word occurs in the document or not. The `CountVectorizer` from the `sklearn.feature_extraction.text` package is used for this purpose with the parameter `ngram_range=(1,2)` so that both unigrams and bigrams are considered.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Create the vectorizer
unibi_counts_vectorizer = CountVectorizer(ngram_range=(1,2))

# Fit training data and transform appropriately
unibi_train_X = unibi_counts_vectorizer.fit_transform(train_set.data)
unibi_train_y = train_set.target

# Transform test data according to learnt training vocabulary
unibi_test_X = unibi_counts_vectorizer.transform(test_set.data)
unibi_test_y = test_set.target
```

```python
print("Length of feature vector: {0}".format(unibi_train_X.shape[1]))
```

###Function Setup

The following code sets up a function to perform classification. It accepts an estimator object and calls the `fit()` method of that object to learn the training data, and then the `predict()` method, supplying the test data as a parameter. The predicted vectors for the test data are returned.

```python
def classify(classifier):
    classifier.fit(unibi_train_X, unibi_train_y)
    predicted_y = classifier.predict(unibi_test_X)
    
    return predicted_y

def select_classify(classifier, k):
    
    # First select
    from sklearn.feature_selection import chi2, SelectKBest
    selector = SelectKBest(chi2, k)
    unibi_seltrain_X = selector.fit_transform(unibi_train_X, unibi_train_y)
    
    classifier.fit(unibi_seltrain_X, unibi_train_y)
    predicted_y = classifier.predict(selector.transform(unibi_test_X))
    
    return predicted_y
```

The code below now sets up a function to calculate accuracy values and report the same for the last performed classification.

```python
def accuracy():
    from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

    acc = accuracy_score(unibi_test_y, unibi_predict_y) * 100
    print("Accuracy: {0}%".format(acc))
    print(classification_report(unibi_test_y, unibi_predict_y, target_names=test_set.target_names))
    
    return acc
```

The code below sets up a function to evaluate a classifier's performance depending on the number of features supplied to it. It creates a plot of features vs. accuracy out of repeated runs of the supplied classifier.

```python
def compare_feature_performance(classifier, classifier_name='Classifier'):
    
    from sklearn.metrics import accuracy_score
    accuracies = []
    kvals = logspace(1, 17, num=17, base=2.0)  # k = 2, 4, 8, 16, ...
    kvals = append(kvals, [unibi_train_X.shape[1]])
    for k in kvals:
        unibi_predict_y = select_classify(classifier, k)
        accuracies.append(accuracy_score(unibi_test_y, unibi_predict_y) * 100)
    
    plot(kvals, accuracies)
    xscale('log')
    xlim([0, unibi_train_X.shape[1]])
    xlabel('Number of features')
    ylabel('Accuracy (%)')
    title('{0} Behaviour with Different no of Features'.format(classifier_name))
```

###Feature Analysis

We now see what features are most indicative of the sentiment.

```python
from sklearn.feature_selection import chi2, SelectKBest
selector = SelectKBest(chi2)
selector.fit(unibi_train_X, unibi_train_y)
best_indices = selector.get_support(indices=True)
feature_names = unibi_counts_vectorizer.get_feature_names()
scores = selector.scores_
max_score = scores.max()
for best_index in best_indices:
    print("{0}\t{1}".format(feature_names[best_index], scores[best_index]/max_score))
```

Here all the features are unigrams. This is because the counts of unigrams are naturally higher than bigrams, and therefore, more indicative as a feature.

###Classifier 5: Stochastic Gradient Descent

####a) Stochastic Gradient Descent with Hinge Loss
This is the loss function used by linear SVM.

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score

# Run classifier 5 times to avoid unpredictable results due to probabilistic nature
accuracy_list = []

for i in range(5):
    unibi_sgd_hinge = SGDClassifier(shuffle=True)
    unibi_predict_y = classify(unibi_sgd_hinge)
    accuracy_list.append(accuracy_score(unibi_test_y, unibi_predict_y) * 100)
```

```python
print("Average accuracy: {0}".format(mean(accuracy_list)))
print("Min accuracy: {0}".format(min(accuracy_list)))
print("Max accuracy: {0}".format(max(accuracy_list)))
unibi_sgd_hinge_accuracy = mean(accuracy_list)
```

```python
from sklearn.linear_model import SGDClassifier

unibi_sgd_hinge = SGDClassifier(shuffle=True)

compare_feature_performance(unibi_sgd_hinge, 'Hinge Loss SGD Classifier')
```

####b) Stochastic Gradient Descent with Modified Huber Loss
This is a smoothed version of hinge loss.

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score

# Run classifier 5 times to avoid unpredictable results due to probabilistic nature
accuracy_list = []

for i in range(5):
    unibi_sgd_huber = SGDClassifier(loss='modified_huber', shuffle=True)
    unibi_predict_y = classify(unibi_sgd_huber)
    accuracy_list.append(accuracy_score(unibi_test_y, unibi_predict_y) * 100)
```

```python
print("Average accuracy: {0}".format(mean(accuracy_list)))
print("Min accuracy: {0}".format(min(accuracy_list)))
print("Max accuracy: {0}".format(max(accuracy_list)))
unibi_sgd_huber_accuracy = mean(accuracy_list)
```

```python
from sklearn.linear_model import SGDClassifier

unibi_sgd_huber = SGDClassifier(loss='modified_huber', shuffle=True)

compare_feature_performance(unibi_sgd_huber, 'Huber SGD Classifier')
```
